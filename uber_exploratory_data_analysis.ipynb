```python
df1["Avg VTAT"]=df1["Avg VTAT"].fillna(df1["Avg VTAT"].mean())
df1['Avg CTAT']=df1['Avg CTAT'].fillna(df1['Avg CTAT'].mean())
```

```python
df1['Date'] = pd.to_datetime(df1['Date'])
df1['Time'] = pd.to_datetime(df1['Time'])

df1['Year'] = df1['Date'].dt.year
df1['Month'] = df1['Date'].dt.month
df1['Day'] = df1['Date'].dt.day_name()
```

```python
plt.figure(figsize=(13,5))
months=df1.groupby('Month').size()
sns.lineplot(x=months.index,y=months.values,marker='o')
plt.title('Number of Rides per Month')
plt.xlabel('Month')
plt.ylabel('Number of Rides')
plt.xticks(ticks=range(1,13),labels=['January','Febrauary','March','April','May','June','July','August','September','October','November','December'],rotation=45)
plt.show()
```

```python
plt.figure(figsize=(13,5))
days=df1.groupby('Day').size().reindex(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
sns.barplot(x=days.index,y=days.values,palette='viridis')
plt.title('Number of Rides per Day')
plt.xlabel('Day')
for x,y in enumerate(days.values):
  plt.text(x,y+10,str(y),ha='center',va='bottom')
plt.ylabel('Number of Rides')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize = (10, 5))
hours = df1.groupby(df1['Time'].dt.hour).size().reindex()

# let's plot a line plot for this

sns.lineplot(x = hours.index, y = hours.values, marker = 'o')
plt.title('Booking by hours of day')
plt.xlabel('Hours')
plt.ylabel('No. of booking')

plt.xticks(range(0, 24))

for i, j in enumerate(hours.values):
    plt.text(i, j+10,str(j), ha='center', va='bottom', fontsize=9, fontweight='bold', rotation = 60 )

plt.show()
```

```python
plt.figure(figsize=(13,5))
vehicle=df1['Vehicle Type'].groupby(df1['Vehicle Type']).size().sort_values(ascending=False)
sns.barplot(x=vehicle.values,y=vehicle.index,palette='viridis')
plt.title('Number of Rides per Vehicle Type')
plt.xlabel('Number of Rides')
plt.ylabel('Vehicle Type')
plt.show()
```

```python
booking = df1.groupby('Booking Status').size()

explode = [0.1 if (v / booking.sum()) >= 0.15 else 0.05 for v in booking.values]


plt.pie(booking.values, labels = booking.index, autopct='%1.1f%%', explode = explode,
    startangle=140, colors = sns.color_palette("deep"), shadow=True,
    wedgeprops={'edgecolor': 'black'})

plt.show()
```

```python
cols = ["Reason for cancelling by Customer", "Driver Cancellation Reason", "Incomplete Rides Reason"]
titles = ["Canceling by Customer", "Canceling by Driver", "Incomplete Rides"]
#2 rows x 2 columns
fig, axes = plt.subplots(2, 2, figsize=(12, 9))  # bigger figure
axes = axes.flatten()  # flatten to 1D array for easy indexing

for i, (col, title) in enumerate(zip(cols, titles)):
    counts = df1[col].value_counts(dropna=True)
    total = counts.sum()

    #Explode slices
    explode = [0.1 if v/total >= 0.15 else 0.05 for v in counts.values]


    wedges, texts, autotexts = axes[i].pie(
        counts,
        labels=counts.index,
        autopct='%1.1f%%',
        startangle=140,
        shadow=True,
        explode=explode,
        wedgeprops={'edgecolor': 'black'},
        colors= sns.color_palette("deep") ,
    )

    for text in texts:
        text.set_fontsize(10)
        text.set_color('black')
    for autotext in autotexts:
        autotext.set_fontsize(11)
        autotext.set_color('white')
        autotext.set_fontweight('bold')

    axes[i].set_title(title, fontsize=15, fontweight='bold')
    axes[i].set_aspect('equal')

#Remove the empty 4th subplot
fig.delaxes(axes[3])

plt.tight_layout()
plt.show()
```
